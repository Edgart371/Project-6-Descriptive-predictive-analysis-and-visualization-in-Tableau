```python


##############################
        
pjdate = pj['Date'].unique()
list(pjdate)
dicdates = {'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}

for y in pj['Date'].tolist():
    if type(y) == str:
        val = [x.lower() for x in y]
        day = val[:2]
        day1 = val[0] + val[1]
        year = val[-2:]
        year1 = year[0] + year[1]
        month = val[3:6]
        month1 = month[0] + month[1] + month[2]
        correctdate = datetime.datetime(2000 + int(year1), dicdates[month1] , int(day1), 0, 0)
        print(y)
        print(correctdate)
        pj["Date"].replace({y: correctdate}, inplace=True)
    else:
        continue
```

```python
#Remove dot from Creditor column and not space
pj['Creditor'] = pj['Creditor'].str.replace(r'[^0-9a-zA-Z:, \n]+', ' ', regex=True)
```

```python
#Convert the float in to interger
pj['Initial Amount'] = pj['Initial Amount'].astype('int64')
```

```python
#Remove time from column with date
pj.at[2,'Time (local)']='00:00:00'
pj.at[71,'Time (local)'] = '00:00:00'
```

```python
# create unique column of ref + amount
#New ietration because whit this unique code I am spleting the same transations but with diferente steps passing in two ou moore banks
pj.columns

pj['str amount'] = pj["Initial Amount"].apply(str)

pj['uniqueref'] = pj['Creditor'] + pj['Reference'] + pj['str amount']

#pj['Rf_Am'] = [' '.join([j, x, str(y), z, i]) for j, x, y , z , i in zip(pj['Creditor'], pj['Reference'], pj['Initial Amount'], pj['Currency'], pj['Bank'])]
```

```python
pj = pj.drop(columns=["str amount"])
```

```python
# further formatting

# collect country code from bank account and add column with country names

countries = []
for row in pj['Bank']:
        if row == 'CHASUS3AXXX': countries.append('USA')
        elif row == 'BINAADADXXX': countries.append('Andorra')
        elif row == 'BSJUARBJXXX': countries.append('Argentina')
        elif row == 'REDJBY22XXX': countries.append('Belarus')
        elif row == 'AGRIFRPPXXX': countries.append('France')
        elif row == 'TSIBAU44XXX': countries.append('Australia')
        elif row == 'FNROCNBQXXX': countries.append('China')
        elif row == 'FINVALTRXXX': countries.append('Albania')
pj['Country']=countries
```

```python
#Convert the date type in to object
pj['Date'] = pj['Date'].astype('object')
```
